```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm
from scipy.optimize import brentq
import pandas as pd

import os, sys
#mpl.font_manager._rebuild()
#sys.path.insert(0, os.environ['thesis_resources'])
plt.style.use(os.path.join(os.environ['thesis_resources'], 'globals', 'thesis.mplstyle'))
import warnings
warnings.filterwarnings('ignore', message='Substituting with a symbol from Computer Modern')

# create color map to identify mf component 
fmax = 5/2
norm = mpl.colors.Normalize(vmin=-fmax, vmax=fmax)
cmap = cm.copper
m = cm.ScalarMappable(norm=norm, cmap=cmap)

from lightshifts.consts import h, hbar, c, eps0
from lightshifts.auxiliary import smart_gen_array, laser_intensity
import lightshifts.lightshift_solver as ls
```

## Calculate potential for the clock states 1S0, 3P0 and for 3P1

** Transitions relevant for clock states and intercombination line light shift **

```python
#import transition data
ls_1S0 = ls('data/atom_yb173.json', 'data/transitions_1S0.json')
ls_3P0 = ls('data/atom_yb173.json', 'data/transitions_3P0.json')
ls_3P1 = ls('data/atom_yb173.json', 'data/transitions_3P1.json')
```

** find magic wavelength **

```python
lambda_m = brentq(lambda w: ls_3P0.scalar_polarizability(w) \
       - ls_1S0.scalar_polarizability(w), 660*1e-9, 800*1e-9)
print('magic wavelength is %1.2f nm'%(lambda_m*1e9))
```

lattice depth there in kHz/(W/cm^2)

```python
print('lattice depth is %1.2f Hz/(W/cm^2)'%ls_1S0.scalar_lightshift(lambda_m))
```

for a given laser beam (L3)

```python
laser_power = 100e-3 #in Watts
beam_waist  = 127e-6 #in meters
l_int = laser_intensity(laser_power, beam_waist)
print('lattice depth for %1.2f W/cm^2 is %1.2f Hz'%(l_int, ls_1S0.scalar_lightshift(lambda_m, l_int)))
```

** SDL polarizability ratio **

```python
lambda_sdl = 670e-9
ls_3P0.scalar_lightshift(lambda_sdl)/ls_1S0.scalar_lightshift(lambda_sdl)
```

**Yellow probe light shift** 

```python
lambda_clock = 578e-9
clock_laser_intensity = laser_intensity(laser_power=50e-3, beam_waist=127e-6)
probe_shift = ls_3P0.scalar_lightshift(lambda_clock, clock_laser_intensity)\
             -ls_1S0.scalar_lightshift(lambda_clock, clock_laser_intensity)

print('yellow probe shift for %1.2f W/cm^2 is %1.2f Hz'%(clock_laser_intensity, 
                                                         probe_shift))
```

### generate light shift table for clock states and 3P1

```python
lam_0 = 200e-9
lam_1 = 1900e-9
n = 400

# generate lightshift data
sing_1S0 = [x['wavelength'] for x in ls_1S0.sorted_transitions_dict(hyperfine=False)]
sing_3P0 = [x['wavelength'] for x in ls_3P0.sorted_transitions_dict(hyperfine=False)]
sing_3P1 = [x['wavelength'] for x in ls_3P1.sorted_transitions_dict(hyperfine=False)]

w_1S0, l_1S0 = smart_gen_array(lambda w: ls_1S0.scalar_lightshift(w), lam_0, lam_1, sing_1S0, n=n)
w_3P0, l_3P0 = smart_gen_array(lambda w: ls_3P0.scalar_lightshift(w), lam_0, lam_1, sing_3P0, n=n)
w_3P1, l_3P1 = smart_gen_array(lambda w: ls_3P1.scalar_lightshift(w), lam_0, lam_1, sing_3P1, n=n)
```

### import Dzuba 2010 data (extracted from eps) for the clock states

```python
# csv files directly copied from dzuba 2010 arxiv eps file
d1 = pd.read_csv('data/dzuba/dzuba1.csv', names=['x','y'])
d2 = pd.read_csv('data/dzuba/dzuba2.csv', names=['x','y'])

#physical boundaries for the d1 data set
x_min = 0
x_max = 0.1*(478.675-99.532)/(437.147-99.532) #physical right data boundary of d1 (from illustrator, rescaled with 0.1 tick position)
y_min = -300 # checked with illustrator
y_max = 800 # checked with illustrator

def rescale_to_physical(x0, y0):
    x_shift = d1.x.min()
    x_scale = (x_max-x_min)/(d1.x.max() - d1.x.min())
    y_shift = d1.y.min()
    y_scale = (y_max-y_min)/(d1.y.max() - d1.y.min())
    
    x = (x0-x_shift)*x_scale+x_min
    y = (y0-y_shift)*y_scale+y_min
    return x,y

dzuba1_x, dzuba1_y = rescale_to_physical(d1.x, d1.y)
dzuba2_x, dzuba2_y = rescale_to_physical(d2.x, d2.y)

# functions to convert a.u. units to SI
def to_wavelength_nm(omega_au):
    au_freq_unit = 1/2.418884326505e-17
    return c/(omega_au*au_freq_unit/(2*np.pi))/1e-9

def to_lightshift_SI(polariz_au):
    """ 
    convert polarizability in a.u. to lightshift in (h Hz)/[1W/cm^2 intensity]. 
    polarizability is energy/(electric field)^2.
    Conversion from Wikipedia.
    """
    polariz_SI = - polariz_au * 4.35974417e-18/(5.14220652e11)**2 /(2*eps0*c) * (1/(0.01)**2/h)
    
    return polariz_SI

# compare with old Mathematica interpolation table (Christian Schweizer)
#dzuba_1S0_old = pd.read_csv('data/tabPolarizability1S0-dzuba2010.csv',names=['w','alpha'])
#dzuba_3P0_old = pd.read_csv('data/tabPolarizability3P0-dzuba2010.csv',names=['w','alpha'])
#plt.plot(dzuba_1S0_old.w.values, to_lightshift_SI(dzuba_1S0_old.alpha.values), ls='--', c='lightgray')
#plt.plot(dzuba_3P0_old.w.values, to_lightshift_SI(dzuba_3P0_old.alpha.values), ls='--', c='lightgray')

#dzuba plot
plt.plot(dzuba2_x, dzuba2_y, '-', c='k', label='1S0')
plt.plot(dzuba1_x, dzuba1_y, '-', c=m.to_rgba(3/2), label='3P0')
#plt.xlim(350,1800)
#plt.ylim(-100, 15)
plt.ylabel(r'$\alpha\, (a.u.)$')
plt.xlabel('$\omega\,(a.u.)$')
plt.legend()
```

```python
# 2018 data from Dzuba email
def import_dzuba(f):
    df = pd.read_csv(f, na_values='nan')
    df.rename(columns={'w(a.u.)':'x', 'alpha0(a.u.)':'y'}, inplace=True)
    df = df.astype(float)
    return df
dzuba_email_d1 = import_dzuba('data/dzuba/dzuba-email-1s0.txt')
dzuba_email_d2 = import_dzuba('data/dzuba/dzuba-email-3p0.txt')
dzuba_email_d3 = import_dzuba('data/dzuba/dzuba-email-3p1.txt')

#dzuba plot
plt.plot(dzuba_email_d1.x, dzuba_email_d1.y, '-', c='k', label='1S0')
plt.plot(dzuba_email_d2.x, dzuba_email_d2.y, '-', c=m.to_rgba(3/2), label='3P0')
plt.plot(dzuba_email_d3.x, dzuba_email_d3.y, '-', c='green', label='3P1')
plt.ylabel(r'$\alpha\, (a.u.)$')
plt.xlabel('$\omega\,(a.u.)$')
plt.legend()
```

```python
plt.figure(figsize=(9,10))
plt.subplot(2,1,1)

plt.text(0.03,0.95,'a',transform=plt.gca().transAxes, weight='bold', size=13)

# our calculation
la, = plt.plot(w_1S0*1e9, l_1S0, 'k', ls='-', label='LS coupling estimate')
lb, = plt.plot(w_3P0*1e9, l_3P0, c=m.to_rgba(3/2), ls='-', label='')

#dzuba plot
plt.plot(to_wavelength_nm(dzuba2_x), to_lightshift_SI(dzuba2_y), '--', c='k', 
         label='CI+MBPT (Dzuba et al.$^a$)')
plt.plot(to_wavelength_nm(dzuba1_x), to_lightshift_SI(dzuba1_y), '--', c=m.to_rgba(3/2), 
         label='')

#dzuba email data plot
plt.plot(to_wavelength_nm(dzuba_email_d1.x.values), 
         to_lightshift_SI(dzuba_email_d1.y), 
         marker='.', ms=2, ls='', mfc='k', c='k', 
         label='CIPT (Dzuba et al.$^b$)')
plt.plot(to_wavelength_nm(dzuba_email_d2.x), 
         to_lightshift_SI(dzuba_email_d2.y), 
         marker='.', ms=2, ls='', mfc=m.to_rgba(3/2), c=m.to_rgba(3/2), 
         label='')

for sing in (sing_1S0, sing_3P0):
    for w in sing:
        plt.plot([w*1e9, w*1e9], [-1e6, 1e6], ':', color='gray')

plt.gca().axhline(y=0, color='k', lw=0.5)
plt.xlim(350,1700)
plt.ylim(-40, 15)
plt.ylabel(r'$\Delta V_\mathrm{ac}/I_0$ [Hz/($W/\mathrm{cm}^2$)]')
plt.xlabel('wavelength (nm)')
leg1 = plt.legend([la, lb], ('${}^1\mathrm{S}_0$', '${}^3\mathrm{P}_0$'), 
                  ncol=2, loc='upper right', 
                  prop={'weight':'bold', 'size':14})
plt.legend()
plt.gca().add_artist(leg1)


## 3P1 lightshift
plt.subplot(2,1,2)
plt.text(0.03,0.95,'b',transform=plt.gca().transAxes, weight='bold', size=13)

la, = plt.plot(w_1S0*1e9, l_1S0, 'k', label='LS coupling estimate')
lb, = plt.plot(w_3P1*1e9, l_3P1, c='green', label='')

plt.plot(to_wavelength_nm(dzuba_email_d1.x), 
         to_lightshift_SI(dzuba_email_d1.y), 
         marker='.', ms=2, ls='', mfc='k', c='k', label='CIPT (Dzuba et al.$^b$)')
plt.plot(to_wavelength_nm(dzuba_email_d3.x), 
         to_lightshift_SI(dzuba_email_d3.y), 
         marker='.', ms=2, ls='', mfc='green',c='green', label='')

for sing in (sing_1S0, sing_3P1):
    for w in sing:
        plt.plot([w*1e9, w*1e9], [-1e6, 1e6], ':', color='gray')

plt.gca().axhline(y=0, color='k', lw=0.5)
plt.xlabel('wavelength (nm)')
plt.ylabel(r'$\Delta V_\mathrm{ac}/I_0$ [Hz/($W/\mathrm{cm}^2$)]')
plt.xlim(350, 1700)
plt.ylim(-50, 50)
plt.grid(False)
leg1 = plt.legend([la, lb], ('${}^1\mathrm{S}_0$', '${}^3\mathrm{P}_1$'), 
                  ncol=2, loc='upper center', 
                  prop={'weight':'bold', 'size':14})
plt.legend()
plt.gca().add_artist(leg1)

#plt.tight_layout()
plt.savefig('figures/polarizabilities_1S0_3P0_3P1.pdf')
```

# OSG ground state polarizability including vector and tensor shifts

### Transitions

```python
# plot range and singularity positions
fzero = ls_1S0.transition_frequency_hyperfine(state_f=['6s6p', '3P1'], Ff=7/2)
lam_0 = c/(fzero+10e9)
lam_1 = c/(fzero-4e9)
n = 200
sing_1S0_OSG = [x['wavelength'] for x in ls_1S0.sorted_transitions_dict(hyperfine=True)]

ls_data_sig = []
for mf in np.arange(-5/2,6/2,1):
    ls_data_sig.append(smart_gen_array(lambda w: ls_1S0.total_lightshift(w, 1, mf), lam_0, lam_1, sing_1S0_OSG, n=100))    

ls_data_pi = []
for mf in np.arange(-5/2,6/2,1):
    ls_data_pi.append(smart_gen_array(lambda w: ls_1S0.total_lightshift(w, 0, mf), lam_0, lam_1, sing_1S0_OSG, n=100))    

data = {'sig': ls_data_sig, 'pi': ls_data_pi}
```

```python
label = {
    'sig': 'a',
    'pi': 'b'
}

title = {
    'sig': '$\sigma^+$ light',
    'pi': '$\pi$ light'
}

legend_title = {
    'sig': '$m_F$',
    'pi': '$|m_F|$'
}

possible_mF = {
    'sig': np.arange(-5/2,6/2,1),
    'pi': np.arange(1/2, 6/2, 1)
}

plt.figure(figsize=(9,4))
for j,pol in enumerate(('sig', 'pi')):
    plt.subplot(1,2,j+1)
    for i,mf in enumerate(possible_mF[pol]):
        w,l = data[pol][i]
        plt.ylim(-100,100)
        plt.plot((c/w-fzero)/1e9, l/1e3, c=m.to_rgba(mf), label=r'{:+1.0f}/2'.format(mf*2))

    for w in sing_1S0_OSG:
        plt.plot([(c/w-fzero)/1e9, (c/w-fzero)/1e9], [-1e6, 1e6], ':', color='gray', label='')
    plt.text(0.05,0.92,label[pol],transform=plt.gca().transAxes, weight='bold', size=13)
    plt.gca().axhline(y=0, color='k', lw=0.5)
    plt.xlim(-4,10)
    plt.ylim(-30,30)
    plt.xlabel(r'detuning from $^1\mathrm{S}_0 \rightarrow ^3\mathrm{P}_1\,(F=7/2)$ (GHz)')
    if j==0:
        plt.ylabel(r'$\Delta V_\mathrm{ac}/I_0$ [kHz/($W/\mathrm{cm}^2$)]')
    plt.legend(loc='upper left',title=legend_title[pol], frameon=False)
    plt.title(title[pol])
plt.savefig('figures/lightshift_OSG.pdf')
```
